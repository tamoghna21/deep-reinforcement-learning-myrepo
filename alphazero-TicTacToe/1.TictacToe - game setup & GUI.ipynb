### Setup a game

```python
# ConnectN is a game environemnt class, that has the states, it keeps track of the player, 1 or -1, and the score.
from ConnectN import ConnectN

game_setting = {'size':(3,3), 'N':3} #3x3 square board
#game_setting = {'size':(3,3), 'N':3, 'pie_rule':True}

game = ConnectN(**game_setting)

print(game.state)
print(game.player)
print(game.score)
```

```python
# A game move is played 
game.move((0,1)) #Player 1 plays a move at square (0,1)
print(game.state) # Game state after the above move
print(game.player) # Now, its turn to the oter player -1
print(game.score)
```

```python
# A few moves are played
game.move((1,1))
game.move((0,2))
game.move((2,1))
game.move((0,0))

print(game.state)
print(game.player)
print(game.score) # game score 1 , that means player 1 won
```
